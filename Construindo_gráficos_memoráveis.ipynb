## O que é o gráfico de linha?

Gráficos de linha são melhores quando você quer mostrar como o valor de algo muda ao longo do tempo, ou comparar como várias coisas mudam ao longo do tempo em relação umas às outras. Sempre que você ouvir essa frase-chave "ao longo do tempo", essa é a sua pista para considerar o uso de um gráfico de linha para seus dados.

Gráficos de linha são gráficos comuns e eficazes porque são simples, fáceis de entender e eficientes. Os gráficos de linha são ótimos para:

* Comparando muitos dados de uma só vez;

* Mostrando mudanças e tendências ao longo do tempo;

* Incluindo contexto importante e anotação;

* Exibindo dados de previsão e incerteza;

* Destacando anomalias dentro e em todas as séries de dados

Por outro lado, eles não são necessariamente a sua melhor escolha para:

* Exibindo quantidades de coisas

* Trabalhando com dados categóricos

* Fazendo comparações de parte a todo

* Mostrando conjuntos de dados em desordem

### Plotando gráfico de linhas na prática


```python
# Configurando a fonte
plt.rcParams['font.family'] = 'Arial'

# Configurar as cores dos gráficos
azul = "#003F5c"
cinza = "#9E9E9E"
cinza_texto = "#515151"
```

```python
# Dados ficticios 
x = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
recebidos = [180, 172, 190, 173, 161, 197, 202, 187, 183, 210, 187, 172]
processados = [180, 170, 184, 169, 159, 187, 192, 180, 169, 171, 140, 124]
```

```python
print("\n\n\n\n")

# Definindo a figura
fig, ax = plt.subplots(figsize=(10, 6), dpi = 72)

# Plots de linhas do gráfico
ax.plot(range(12), recebidos, color = cinza, linewidth = 4)
ax.plot(range(12), processados, color = azul, linewidth = 4)

# Plot dos scatter plots
ax.scatter(range(12)[-3:], recebidos[-3:], s = 90, color = cinza, clip_on = False)
ax.scatter(range(12)[-3:], processados[-3:], s = 90, color = azul, clip_on = False)

# Anotações de números nos pontos marcados
for i, j in enumerate(recebidos[-3:]):
  ax.annotate(
      j,
      xy = (i+9, j),
      xytext = (5, 5),
      textcoords = 'offset points',
      color = cinza_texto,
      fontsize = 12,
      fontweight = 'medium'
      )
for i, j in enumerate(processados[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=azul,
        fontsize=12,
        fontweight='medium'
        )

# Legenda no final de cada linha
for i, j in enumerate(recebidos[-3:]):
  ax.annotate(
      "Recebidos",
      xy=(len(recebidos)-1, recebidos[-1]),
      xytext=(50,0),
      textcoords = 'offset points',
      color = cinza,
      fontsize = 18,
      fontweight = 'semibold'
      )

ax.annotate(
    "Processados",
    xy=(len(recebidos)-1, processados[-1]),
    xytext=(55,0),
    textcoords='offset points',
    color=azul,
    fontsize=18,
    fontweight='semibold'
    )

# Remover spines da direira e superior

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Mudar range do eixo y
ax.set_ylim(100, 220)

# Remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# Configurar xticks - subsituir números por meses (12 meses)
ax.set_xticks(ticks = range(12));
ax.set_xticklabels(labels = x);

# title
ax.text(-0.5, 235,'Taxa de vendas', fontsize=24, color=cinza_texto, fontweight='bold')
ax.text(-0.5, 228,'Quantidade de produtos recebidos e processados', fontsize=13, color=cinza)


# Rodapé
ax.text(
    -0.5,
    80,
    "Fonte: Gráfico construído pelo Professor Thiago Nunes"
)

# Exibir gráfico
plt.show()
```

## O que é gráfico de barras?

Um gráfico de barras é um tipo de gráfico comum para gráficos de dados categóricos ou dados classificados em grupos. Consiste em vários retângulos alinhados a uma linha de base comum. O comprimento de cada um é proporcional ao valor que ele representa — em outras palavras, em um gráfico de barras, os dados são codificados por comprimento. Nossos olhos são muito bons em comparar comprimentos quando os objetos estão alinhados, tornando este gráfico fácil de interpretar — apenas uma razão pela qual os gráficos de barras são comuns.

Nossos olhos começam na base e varrem para o final de cada barra. Medimos os comprimentos relativos tanto à linha de base quanto às outras barras, por isso é um processo simples para identificar o menor ou o maior bar. Também podemos ver o espaço negativo entre diferentes alturas das barras para comparar a diferença incremental entre elas. 

Esses gráficos são fáceis de ler, mas também são amplamente reconhecidos. As chances são de que você já tenha encontrado um gráfico de barras horizontal ou vertical padrão. Mas as barras vêm em muitas formas e tamanhos. Vou listar abaixo algumas das variações mais comuns, com links para exemplos.

### Plotando gráfico de barras na prática

```python
# Dados fictícios

# Os dados do dataset:
tamanho = [7, 5, 8, 61, 45]
barras = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(barras))
```

```python
# Definindo a figura
fig, ax = plt.subplots(figsize=(10, 6), dpi = 72)

# Configurando a fonte
plt.rcParams['font.family'] = 'Arial'

# Configurar as cores do gráfico
cores = ["#9E9E9E", "#9E9E9E", "#9E9E9E", '#003F5c', "#9E9E9E"]
azul = "#003F5c"
cinza = "#9E9E9E"
cinza_texto = "#515151"

# Remover spines da direira e superior

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Mudar range do eixo y
ax.set_ylim(0, 70)

# Remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

ax.bar(barras, tamanho, color=cores);

# title
ax.text(-0.65, 80,'Recorrência de vendas', fontsize=24, color=cinza_texto, fontweight='bold')
ax.text(-0.65, 75,'Análise dos 5 produtos da loja', fontsize=13, color=cinza)

ax.text(-0.24, 10,'14,3 %',fontsize=16, fontweight='bold', color='#9E9E9E')
ax.text(0.730, 7,'16,7 %',fontsize=16, fontweight='semibold', color='#9E9E9E')
ax.text(1.755, 10,'0,0 %',fontsize=16, fontweight='semibold', color='#9E9E9E')
ax.text(2.700, 63,'44,4 %',fontsize=16, fontweight='semibold', color= '#003F5c')
ax.text(3.700, 47,'24,6 %',fontsize=18, fontweight='semibold', color='#9E9E9E')


# Rodapé
ax.text(
    -0.65,
    -10,
    "Fonte: Gráfico construído pelo Professor Thiago Nunes"
)

# Exibir gráfico
plt.show()
```
